```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
receive = [160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177]
processed = [160 ,184 ,237 ,148 ,181 ,150 ,123 ,156 ,126 ,104 ,124 ,140]
```

```python
plt.plot(months, receive,color='#00bebe',label='Recebidos',)
plt.plot(months, processed,color='#be0000', label='Processados')
plt.ylim(90, 250)
plt.title('Número de Processos Recebidos e Processados por Mês',fontdict={'fontsize': 15})
plt.legend()
# plt.show()
plt.savefig('processos.png')
```
